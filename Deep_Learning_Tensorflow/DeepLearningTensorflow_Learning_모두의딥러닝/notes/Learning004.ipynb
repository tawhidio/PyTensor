# 광물예측(과적합 피하기)

```python
# import tensorflow as tf
import pandas as pd
import numpy as np

from keras.models import Sequential, load_model
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
# from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold


# 시드값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)

# 데이터 읽기
ds = pd.read_csv('dataset/sonar.csv', header=None)
print(ds.head())
print(ds.info())
print(ds)
```

```python
# 데이터 입력
dataset = ds.values
x = dataset[:,0:60]
y_obj = dataset[:,60]

# 문자열변환
e = LabelEncoder()
e.fit(y_obj)
y = e.transform(y_obj)
```

```python
#  학습셋(70%)과 테스트셋(30%)
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=seed)

# 10개의 파일로 쪼갬(10겹 교차 검증)
n_fold = 10
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)

# 처리값을 담기위한 배열 생성
accuracy = []
loss = []
```

```python
for train, test in skf.split(x, y):
    # 모델 설정 
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # 모델 컴파일 
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

    # 모델 실행
    model.fit(x[train], y[train], epochs=100, batch_size=5)
    k_loss = "%.4f"%(model.evaluate(x[test], y[test])[0])
    k_accuracy = "%.4f"%(model.evaluate(x[test], y[test])[1])
    
    accuracy.append(k_accuracy)
    loss.append(k_loss)
    
```

```python
# 결과출력
print("%.f fold "%(n_fold))
print("성공률 : ", accuracy, "\n")
print("실패율: ", loss ,"\n")

# 모델 저장 경로및 파일명
file_full_n_p = './save_models/test_model001.h5'

# 모델 저장
model.save(file_full_n_p)
print("save model....")

# 기존 모델 메모리에서 삭제
del model

# 모델 다시 불러오기
model = load_model(file_full_n_p)
```
