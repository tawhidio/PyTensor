# 그래프 라이브러리 사용

* ploat() 함수
  * 직선 또는 꺽은선 형태의 그래프를 그릴 때 사용한다.
  * plot(X축값, Y축값)
  
  1. 직선 
     * Y축 값으로 입력 되면  X축 값은 자동으로 1씩 증가한다.

```python
import matplotlib.pyplot as plt
import csv

plt.title('YourStock Plotting')
plt.plot([10, 20, 30, 40])

plt.show()
```


  2. 꺽은선
     * X 축 <code>[1, 2, 3, 4]</code> 까지 증가시키고 <code>[10, 20, 20, 30]</code>값으로 처리

```python
plt.title('MyStock Plotting')
plt.plot([1, 2, 3, 4], [30, 40, 10, 5])

plt.show()
```

   3. 그래프에 범례 넣기
      * label을 사용하여 표시. label='범례'
      * 범례를 표시하기위해서는 plt.show()를 하기전에 legend() 함수를 호출해준다. plt.legend()

```python
plt.title('Stock Ploting')

plt.plot([10, 20, 30, 40], label='YourStock')
plt.plot([1, 2, 3, 4], [30, 40, 10, 5], label='MyStock')
plt.plot([1, 2, 3, 4], [15, 30, 10, 20], label='NomalStock')

plt.legend()
plt.show()
```

   4. 라인 선 색깔및 선 모양 바꾸기
      * color로 색깔  변경가능(#000000 도 가능)
      * linestyle로 라인 모양 변경가능

```python
plt.title('EveryStock')
plt.plot([1, 2, 3, 4, 5, 6], [20, 10, 30, 60, 40, 50], color='pink', linestyle="--", label='HapayThink')
plt.legend()
plt.show()
```

   5. 마커모양 바꾸기
      * 'r.' 원형으로 표시
      * 'g^' 삼각형으로 표시

```python
plt.title('ChangeMaker')

plt.plot([1, 2, 3, 4, 5, 6, 7], [6, 2, 5, 7, 1, 3, 3], 'r.', label='circle')
plt.plot([1, 2, 3, 4, 5, 6, 7], [4, 2, 3, 1, 4, 6, 7], 'g^', label='triangle up')
plt.legend(loc=5)
plt.show()
```
