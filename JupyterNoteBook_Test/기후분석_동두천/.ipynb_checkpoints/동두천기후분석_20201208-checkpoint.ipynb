```python
import csv                                               # csv 함수 호출
f = open('ta_20201208150322.csv', 'r', encoding='cp949') # csv파일을 읽기 전용으로 인코딩은 cp949로 
data = csv.reader(f, delimiter=',')                      # csv를 읽어들이는데 구분자는 ',' 로 한다.
for row in data :                                       # data 값을 row에 치환하여 반복
    print(row)                                           # 출력
f.close()                                               # 파일 모두 출력후 파일닫기
```
